## Text analysis with nltk

```python
import nltk
```

```python
f = open('./txt/text_example.txt', 'r')
text = f.read()
f.close()

print(text)
```

```python
cleaned_text = text.replace('[','').replace(']','').replace(',','').replace('.','').replace('"','').replace('\n', ' ')

# Case conversion (from upper cases to lower cases)
cleaned_text = cleaned_text.lower()
print(cleaned_text)
```

```python
word_tokens = nltk.word_tokenize(cleaned_text)
print(word_tokens)
```

```python
# POS tagging
tokens_pos = nltk.pos_tag(word_tokens)
tokens_pos
```

```python
noun_words = [word for word, pos in tokens_pos if pos.find('NN') >= 0]
print(noun_words)
```

## Web page analysis (with. The new york times)

```python
import requests
from bs4 import BeautifulSoup

def get_article_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    title = soup.title.text
    contents = soup.findAll('p', class_='g-body')
    content_text = ''
    for content in contents:
        content_text = content_text + ' ' + content.text

    return title, content_text

url = 'https://www.nytimes.com/interactive/2021/05/28/climate/climate-wind-solar-energy-map.html'

title, content = get_article_content(url)
print(title, content)
```

```python
symbols = ['“', "“", ',', '.', ':']

for symbol in symbols:
    content = content.replace(symbol, '')
```

```python
content = content.lower()
```

```python
content
```

```python
tokens = nltk.word_tokenize(content)
```

```python
pos_tokens = nltk.pos_tag(tokens)
```

```python
pos_tokens
```

```python
noun_words = [word for word, pos in pos_tokens if pos.find('NN') >= 0]
print(noun_words)
```

```python
customized_stopwords = ['s', 'ms', "'"]
unique_noun_words = set(noun_words)
for word in unique_noun_words:
    if word in customized_stopwords:
        while word in noun_words: noun_words.remove(word)

print(noun_words)
```
